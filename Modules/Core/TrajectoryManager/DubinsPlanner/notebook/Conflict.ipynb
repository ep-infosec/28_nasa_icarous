```
from sympy import *
init_printing()
```

```
x1,y1,z1 = symbols('x_1,y_1,z_1')
x2,y2,z2 = symbols('x_2,y_2,z_2')
xa,ya,za = symbols('x_a,y_a,z_a')
xb,yb,zb = symbols('x_b,y_b,z_b')
vxa,vya,vza = symbols('v_x_a,v_y_a,v_z_a')
vxb,vyb,vzb = symbols('v_x_b,v_y_b,v_z_b')
t,t1,t2 = symbols('t,t_1,t_2')
D = symbols('D')
```

```
x1 = xa + vxa*(t-t1)
x2 = xb + vxb*(t-t2)
y1 = ya + vya*(t-t1)
y2 = yb + vyb*(t-t2)
z1 = za + vza*(t-t1)
z2 = zb + vzb*(t-t2)
```

```
xr = x2 - x1
yr = y2 - y1
p1 = xr**2 + yr**2 - D**2
display(p1)
ans = solve(p1,t)
sub = {xa:10,xb:20,ya:0,yb:0,t1:50,t2:0,vxa:1,vya:0,vxb:-1,vyb:0,D:0}
t1_0 = ans[0].subs(sub)
t2_0 = ans[1].subs(sub)
display(t1_0.evalf(),t2_0.evalf())

_x1 = x1.subs(sub).subs({t:t1_0})
display(x2.subs(sub))
_x2 = x2.subs(sub).subs({t:t1_0})
display(_x1.evalf(),_x2.evalf())
```

```
p2 = expand(p1)
display(p2)
```

```
(p2.coeff(t**2))*t**2 + (p2.coeff(t))*t + (p2.coeff(t,n=0))
```

```
a = p2.coeff(t**2)
b = p2.coeff(t)
c = p2.coeff(t,n=0)
display(a); display(b); display(c)
print("b=",b)
print("c=",c)
```

```
p3 = expand(x1**2 + y1**2 - D**2)
a = p3.coeff(t**2)
b = p3.coeff(t)
c = p3.coeff(t,n=0)
display(a); display(b); display(c)
```
